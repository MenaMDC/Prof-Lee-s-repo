```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib


# Read in the image
image_apollo11 = cv2.imread('/content/Apollo-11-launch.jpg', cv2.IMREAD_COLOR)

green = (0, 255, 0)       # Green in BGR
magenta = (255, 0, 255)   # Magenta in BGR

# Create a copy
image_annotated = image_apollo11.copy()

# Add text to the dark bottom area
text = 'Apollo 11 Saturn V Launch, July 16, 1969'
font_face = cv2.FONT_HERSHEY_PLAIN
font_scale = 2
font_color = green
font_thickness = 2

cv2.putText(image_annotated, text, (300, 700), font_face, font_scale, font_color, font_thickness, cv2.LINE_AA)

# Draw a magenta rectangle
cv2.rectangle(image_annotated, (500, 100), (700, 650), magenta, thickness=3, lineType=cv2.LINE_8)

# Display the final image
plt.figure(figsize=[8, 8])
plt.imshow(image_annotated[:, :, ::-1])
```
